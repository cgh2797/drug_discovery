```python
import sys
sys.path.append('/usr/local/lib/python3.7/site-packages/')

from rdkit.Chem import MCS
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Descriptors
from rdkit.Chem import AllChem
from rdkit import DataStructs
from rdkit.Chem import rdFMCS
from rdkit.Chem import rdMolDescriptors
import timeit

import numpy as np
import pandas as pd
import seaborn as sns


#descriptor
from rdkit.Chem.Descriptors import ExactMolWt
from rdkit.Chem.Crippen import MolLogP
from rdkit.Chem.rdMolDescriptors import CalcNumHBD
from rdkit.Chem.rdMolDescriptors import CalcNumHBA
from rdkit.Chem.rdMolDescriptors import CalcTPSA

import matplotlib.pyplot as plt
```

```python
realfirst=timeit.default_timer()
start1 = timeit.default_timer()
data=pd.read_csv('CID_smiles_44.csv')
stamp1= timeit.default_timer()
print('파일읽기 실행시간은 : ')
print(stamp1- start1)
```

```python
start1 = timeit.default_timer()
smiles_list=data['smiles']
stamp1 = timeit.default_timer()
print('데이터 추출 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
mol_list=[]
for smiles in smiles_list:
    mol=Chem.MolFromSmiles(smiles)
    mol_list.append(mol)
stamp1 = timeit.default_timer()
print('mol변환 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
result=[]
for i in  range(len(mol_list)) :
    try:
        MolLogP(mol_list[i])
        result.append(True)
    except:
        result.append(False)
stamp1 = timeit.default_timer()
print('error 잡는 실행시간은 : ')
print(stamp1-start1)
```

```python
data['result']=result
data1=data[result]
data2=data1.reset_index(drop=True)
smiles_list2=data2['smiles']
```

```python
start1 = timeit.default_timer()
MW=[]
for smiles in mol_list2:
    mol=ExactMolWt(smiles)
    MW.append(mol)
stamp1 = timeit.default_timer()
print('MW계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
LogP=[]
for smiles in mol_list2:
    mol=MolLogP(smiles)
    LogP.append(mol)
stamp1 = timeit.default_timer()
print('LogP계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
HBD=[]
for smiles in mol_list2:
    mol=CalcNumHBD(smiles)
    HBD.append(mol)
stamp1 = timeit.default_timer()
print('HBD계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
HBA=[]
for smiles in mol_list2:
    mol=CalcNumHBA(smiles)
    HBA.append(mol)
stamp1 = timeit.default_timer()
print('HBA계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
TPSA=[]
for smiles in mol_list2:
    mol=CalcTPSA(smiles)
    TPSA.append(mol)
stamp1 = timeit.default_timer()
print('TPSA계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
HBAD=[]
for smiles in mol_list2:
    mol=CalcNumHBD(smiles)+CalcNumHBA(smiles)
    HBAD.append(mol)
stamp1 = timeit.default_timer()
print('HBA+HBD계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
rota=[]
for smiles in mol_list2:
    mol=rdMolDescriptors.CalcNumRotatableBonds(smiles)
    rota.append(mol)
stamp1 = timeit.default_timer()
print('rotatable bonds계산 실행시간은 : ')
print(stamp1-start1)
```

```python
start1 = timeit.default_timer()
data2['MW']=MW
data2['LogP']=LogP
data2['HBA']=HBA
data2['HBD']=HBD
data2['TPSA']=TPSA
data2['HBA_HBD']=HBAD
data2['rotational bond']=rota
stamp1=timeit.default_timer()
print('데이터 합치기 실행시간은 : ')
print(stamp1-start1)
```

```python
data2.drop('result',axis=1,inplace=True)
```
