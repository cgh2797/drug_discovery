```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import PandasTools

import timeit
start1 = timeit.default_timer()
my_sdf_file = 'Compound_000000001_000500000.sdf'



frame = PandasTools.LoadSDF(my_sdf_file,
                            smilesName='SMILES',
                            molColName='Molecule',
                             includeFingerprints=False)


stamp1 = timeit.default_timer()



print('parsing 실행시간은 : ')
print(stamp1-start1)
```

```python

start2 = timeit.default_timer()
data=frame[['PUBCHEM_COMPOUND_CID','PUBCHEM_CACTVS_COMPLEXITY','PUBCHEM_CACTVS_HBOND_ACCEPTOR',
           'PUBCHEM_CACTVS_HBOND_DONOR','PUBCHEM_CACTVS_ROTATABLE_BOND','PUBCHEM_XLOGP3_AA',
            'PUBCHEM_MOLECULAR_WEIGHT','PUBCHEM_OPENEYE_CAN_SMILES','PUBCHEM_CACTVS_TPSA','PUBCHEM_TOTAL_CHARGE' 
           ]]
stamp2= timeit.default_timer()
print('데이터 필터링 실행시간은 : ')
print(stamp2- start2)
```

```python
start3 = timeit.default_timer()
data.to_csv('test.csv')
stamp3= timeit.default_timer()
print('파일저장 실행시간은 : ')
print(stamp3- start3)
```
