```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
import numpy as np
```

```python
df_excel = pd.read_excel('pubchem_cid_tinib_795.xlsx') 
```

```python
CID=df_excel['CID']
```

```python
smiles=[]
q=0
print(str(len(CID)) +' jobs need \ntask start')
for i in CID:
    url='https://pubchem.ncbi.nlm.nih.gov/compound/'
    driver = webdriver.Chrome("C:/Users/chris/pubchem crwaling/chromedriver.exe")
    url=url+str(i)
    driver.get(url)
    time.sleep(4)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    driver.quit()
    time.sleep(2)
    try:
        data=soup.select('#Canonical-SMILES > div.section-content > div.section-content-item > p')[0].get_text()
    except:
        data='err'
    smiles.append(data)
    print(str(q)+'th job finished')
    q=q+1
    time.sleep(2)
```

```python
smiles
```

```python
df_excel['smiles']=smiles
```

```python
df_excel.to_excel('795_CID_SMILES_name.xlsx', index=False)
```
